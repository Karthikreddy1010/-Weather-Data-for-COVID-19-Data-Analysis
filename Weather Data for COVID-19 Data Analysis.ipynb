### Null values state, min,max,slp,dewp,rh,ah
### I replace null values with median value because min,max,slp,dewp,rh,ah contains outliers

```python
def replace_null(data, cols):
    for col in cols:
        if data[col].dtype in ['int64', 'float64']:  
            median = data[col].median()             
            data[col].fillna(median, inplace=True)  
        else:
            mode = data[col].mode()[0]              
            data[col].fillna(mode, inplace=True)    

cols = ['Province/State', 'min', 'max', 'dewp', 'rh', 'ah','slp']

replace_null(data, cols)
data.isnull().sum()
```

### Null values are replaced by median and mode

# 1) How do confirmed cases and fatalities vary over time for different regions?

```python
data.head(1)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.lineplot(data=data,x='day_from_jan_first',y='ConfirmedCases')
plt.title("Confirmed Cases over the time")
plt.xlabel("On Jan 1st Cases")

plt.subplot(1,2,2)
sns.lineplot(data=data,x='day_from_jan_first',y='Fatalities')
plt.title("fatalities over the time")
plt.xlabel("On Jan 1st Fatalities")
```

```python
sns.lineplot(data,x='day_from_jan_first',y='ConfirmedCases',label='Confirmed cases')
sns.lineplot(data=data,x='day_from_jan_first',y='Fatalities',label='Fatalities')
plt.legend()
plt.title('Trends of Confirmed Cases and Fatalities on Jan 1st')
plt.show()
```

```python
# Aggregate data by region
region_data = data.groupby('Country/Region')[['ConfirmedCases', 'Fatalities']].sum().reset_index()
region_data = region_data.sort_values(by='ConfirmedCases', ascending=False)

plt.figure(figsize=(10, 28))
sns.barplot(data=region_data, x='ConfirmedCases', y='Country/Region', color='blue', label='Confirmed Cases')
sns.barplot(data=region_data, x='Fatalities', y='Country/Region', color='red', label='Fatalities')
plt.title('Total Confirmed Cases and Fatalities by Region')
plt.legend()
plt.show()
```

# 2) Which countries or regions have the highest and lowest total confirmed cases and fatalities?


```python
regional_totals = data.groupby('Country/Region')[['ConfirmedCases', 'Fatalities']].sum().reset_index()

highest_cases = regional_totals.sort_values(by='ConfirmedCases', ascending=False).head(10)
lowest_cases = regional_totals[regional_totals['ConfirmedCases'] > 0].sort_values(by='ConfirmedCases').head(10)

highest_fatalities = regional_totals.sort_values(by='Fatalities', ascending=False).head(10)
lowest_fatalities = regional_totals[regional_totals['Fatalities'] > 0].sort_values(by='Fatalities').head(10)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=highest_cases,x='Country/Region',y='ConfirmedCases',color='blue',label='ConfirmedCases')
sns.barplot(data=highest_fatalities,x='Country/Region',y='Fatalities',color='red',label='Fatalities')
plt.title("Highest Cases and fatalities by Countries")
plt.legend()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=lowest_cases,x='ConfirmedCases',y='Country/Region',color='blue',label='ConfirmedCases',orient='h')
sns.barplot(data=lowest_fatalities,x='Fatalities',y='Country/Region',color='red',label='Fatalities',orient='h')
plt.title("Lowest Cases and fatalities by Countries")
plt.legend()
```
